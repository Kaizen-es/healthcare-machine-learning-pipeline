```python
import pandas as pd
import numpy as np
import plotly.express as px
import kagglehub
import os
```

```python
path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")


```

```python
# import data
df = pd.read_csv(path + "/data.csv")
pd.set_option('display.max_columns', None)
# data exploration
df.head()
df.columns
df.shape
```

```python
print(df['Unnamed: 32'])
#print(df.describe())
df.info()
```

```python
df.duplicated().sum()
df.isnull().sum()
```

```python
# Data Cleaning
df_clean = pd.read_csv(path + "/data.csv")
df_clean.drop(['Unnamed: 32','id'], axis=1, inplace=True)
df_clean.columns
df_clean.shape
```

```python
# Data visualization


px.histogram(df_clean, x ='radius_mean', color= 'diagnosis', opacity= 0.8, labels={'M':'Malignant'}, barmode= 'overlay', title='Distribution of radius_mean').show()
px.histogram(df_clean, x ='texture_mean', color= 'diagnosis', title='Distribution of texture_mean').show()
px.histogram(df_clean, x ='perimeter_mean', title='Distribution of perimeter_mean').show()
px.histogram(df_clean, x ='area_mean', title='Distribution of area_mean').show()
px.histogram(df_clean, x ='perimeter_se', title='Distribution of perimeter_se').show()
px.histogram(df_clean, x ='perimeter_worst', title='Distribution of perimeter_worst').show()
px.pie(df_clean, names='diagnosis', hole=0.4, title='Class Distribution')

```
